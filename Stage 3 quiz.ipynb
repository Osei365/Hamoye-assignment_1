```python
#Normalising the quiz data

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

quiz_df = quiz_data.drop(['date', 'lights'], axis =1)

normalised_quiz_df = pd.DataFrame(scaler.fit_transform(quiz_df), columns = quiz_df.columns)
normalised_quiz_df
```

```python
# Import train_test_split to split the dataset into train and test
from sklearn.model_selection import train_test_split
X = normalised_quiz_df.drop('Appliances', axis = 1)
y = normalised_quiz_df['Appliances']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```
