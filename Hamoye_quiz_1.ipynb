```python
import pandas as pd
from scipy.stats import skew, kurtosis

data = pd.read_csv('electricity_data.csv')
```

```python
data
```

```python
data.describe()[['fuel_mmbtu_per_unit']].loc[['std', '75%'], 'fuel_mmbtu_per_unit'].reset_index()
```

#### The std and 75% percentile is 10.60 and 17.01 rspectively

```python
data.groupby('fuel_type_code_pudl')['fuel_cost_per_unit_burned'].mean().reset_index().sort_values('fuel_cost_per_unit_burned')
```

#### From the code, the fuel type code with the lowest average fuel cost per unit burned is "gas"

```python
print(skew(data['fuel_qty_burned']))
print()
print(kurtosis(data['fuel_qty_burned']))
```

#### The skew and kurtosis values are 15.85 and 651.26 rsepctively

```python
data.isnull().sum()
```

```python
(data.isnull().sum()['fuel_unit']/len(data)) * 100
```

#### The column with missing values is "fuel_unit", while the number of missing values are "180" and the percentage is "0.610"

```python
data.corr()['fuel_cost_per_unit_burned'].reset_index().sort_values('fuel_cost_per_unit_burned')
```

#### The feature with the second and third lowest correlation with fuel_cost_per_unit_burned is "fuel_qty_burned" and "fuel_mmbtu_per_unit" respectively

```python
year_groupby = data.groupby('report_year')['fuel_cost_per_unit_delivered'].mean().reset_index()
year_groupby.sort_values('fuel_cost_per_unit_delivered', ascending = False)
```

#### From the result above, the year with the highest average fuel cost per unit learnt is "1997"

```python
fuel_type = data.groupby(['fuel_type_code_pudl','report_year'])['fuel_cost_per_unit_burned'].sum().reset_index()
fuel_type = fuel_type.set_index('report_year')
fuel_type = fuel_type[fuel_type['fuel_type_code_pudl'] == 'coal']
fuel_type
```

```python
year_1994 = fuel_type['fuel_cost_per_unit_burned'][1994] 
year_1998 = fuel_type['fuel_cost_per_unit_burned'][1998]

((year_1994 - year_1998)/year_1994) * 100
```

#### The percentage change in fuel cost per unit burned in 1998 compared to 1994 for coal fuel type is......"21%"
